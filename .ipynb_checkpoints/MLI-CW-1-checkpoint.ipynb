```python
#imports
import sklearn
from sklearn.datasets import make_moons
from sklearn.preprocessing import PolynomialFeatures 
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier, AdaBoostClassifier
import numpy as np
import matplotlib.pyplot as plt
```

1) Create a dataset of 1000 points with a noise value of 0.1 and random_state=0 and augment it using polynomial features (recommended degree 10-15) (4 points)

```python
## ADD YOUR CODE HERE
dataset, dataset_labels = make_moons(n_samples=1000, shuffle=True, noise=0.45, random_state=0)
poly = PolynomialFeatures(10, include_bias=False)
dataset = poly.fit_transform(dataset)

split_index = int(len(dataset)*0.9)
dataset_train, dataset_test = dataset[:split_index], dataset[split_index:]
dataset_labels_train, dataset_labels_test = dataset_labels[:split_index], dataset_labels[split_index:]
```

2) Visualize your input data - the first 2 dimensions would suffice  (3 points)



```python
## ADD YOUR CODE HERE
color = []
for label in dataset_labels:
    if label == 1:
        color.append('r')
    else:
        color.append('g')
plt.scatter(dataset[:,0], dataset[:,1], color=color)
plt.show()
```

```python
def plot_decision_boundary(model):
    xy = np.mgrid[-2:3:0.01, -1.5:2.5:0.01].reshape(2,-1).T
    poly = PolynomialFeatures(10, include_bias=False)
    dataset = poly.fit_transform(xy)
    labels = model.predict_proba(dataset)
    colors = np.zeros((len(labels), 4))
    for i,label in enumerate(labels):
        if label[0]> label[1]:
            colors[i][1] = 1.0
            colors[i][3] = label[0] 
        else:
            colors[i][0] = 1.0
            colors[i][3] = label[1]
    plt.scatter(xy[:,0], xy[:,1], color=colors, s=1)
    plt.show()
```

3) Use a Decision Tree to classify the data (5 points)



```python
## ADD YOUR CODE HERE
model = DecisionTreeClassifier(criterion='gini', max_depth=None, random_state=0)
model.fit(dataset_train, dataset_labels_train)
print(model.score(dataset_test, dataset_labels_test))

```

4) Try both Gini Impurity and Information Gain, which performs better? What is the difference between them? (4 points)

```python
## Gini index
model = DecisionTreeClassifier(criterion='gini', max_depth=None, random_state=0)
model.fit(dataset_train, dataset_labels_train)
print(model.score(dataset_test, dataset_labels_test))
plot_decision_boundary(model)
## Gain information
model = DecisionTreeClassifier(criterion='entropy', max_depth=None, random_state=0)
model.fit(dataset_train, dataset_labels_train)
print(model.score(dataset_test, dataset_labels_test))
plot_decision_boundary(model)
```

5) Perform classification using a tree depth of 1-20. Plot the accuracy versus tree depth. (5 points)



```python
## ADD YOUR CODE HERE
depths = [depth for depth in range(1,21)]
accuracies = []
for depth in depths:
    model = DecisionTreeClassifier(criterion='gini', max_depth=depth, random_state=0)
    model.fit(dataset_train, dataset_labels_train)
    accuracies.append(model.score(dataset_test, dataset_labels_test))
plt.plot(depths, accuracies)
```

7) Use Bootstrap Classifier to classify your data, base estimator should be a decision tree. Tune hyperparameters of tree depth, n_of estimators, max_samples and bootstrap. Plot accuracy versus different hyperparameters. (7 points)



```python
## ADD YOUR CODE HERE
### Initial parameters 
max_depth = None
max_samples = 1.0
n_estimators = 10
bootstrap = True
    
### Look for best depth
depths = [depth for depth in range(1,41)]
accuracies = []

for depth in depths:
    ### Initialization
    bagging = BaggingClassifier(
        DecisionTreeClassifier(criterion='entropy', max_depth=depth, random_state=0), 
        max_samples=max_samples, bootstrap=bootstrap, n_estimators=n_estimators, random_state=0, n_jobs=-1)

    ### Evaluation
    bagging.fit(dataset_train, dataset_labels_train)
    accuracies.append(bagging.score(dataset_test, dataset_labels_test))

## Plot
fig, ax = plt.subplots()
ax.plot(depths, accuracies)
ax.set(xlabel='depth', ylabel='accuracy', title='Accuracy for different estimator depth')
ax.grid()

## Best depth
max_depth = 15
```

```python
### Look for best n_estimator
estimators = [n_estimator for n_estimator in range(1, 51)]
accuracies = []

for n_estimators in estimators:
    ### Initialization
    bagging = BaggingClassifier(
        DecisionTreeClassifier(criterion='entropy', max_depth=max_depth, random_state=0), 
        max_samples=max_samples, bootstrap=bootstrap, n_estimators=n_estimators, random_state=0, n_jobs=-1)

    ### Evaluation
    bagging.fit(dataset_train, dataset_labels_train)
    accuracies.append(bagging.score(dataset_test, dataset_labels_test))

## Plot
fig, ax = plt.subplots()
ax.plot(estimators, accuracies)
ax.set(xlabel='n_estimator', ylabel='accuracy', title='Accuracy for different number of estimators')
ax.grid()

## Best n_estimators
n_estimators = 10

```

```python
### Look for best max_samples
max_samples = [n_estimator for n_estimator in np.arange(0.02, 1.02, 0.02)]
accuracies = []
for max_sample in max_samples:
    ### Initialization
    bagging = BaggingClassifier(
        DecisionTreeClassifier(criterion='entropy', max_depth=max_depth, random_state=0), 
        max_samples=max_sample, bootstrap=bootstrap, n_estimators=n_estimators, random_state=0, n_jobs=-1)
    
    ### Fit
    bagging.fit(dataset_train, dataset_labels_train)
    
    ### Evaluation
    accuracies.append(bagging.score(dataset_test, dataset_labels_test))

## Plot
fig, ax = plt.subplots()
ax.plot(max_samples, accuracies)
ax.set(xlabel='max_sample', ylabel='accuracy', title='Accuracy for different max_sample values')
ax.grid()

## Best max_samples
max_samples = 1.0

```

```python
###Accuracy with bootstrap 
bootstrap = True
bagging = BaggingClassifier(
        DecisionTreeClassifier(criterion='entropy', max_depth=max_depth, random_state=0), 
        max_samples=max_samples, bootstrap=bootstrap, n_estimators=n_estimators, random_state=0)
bagging.fit(dataset_train, dataset_labels_train)
print("Accuracy with bootstrap: {}".format(bagging.score(dataset_test, dataset_labels_test)))
plot_decision_boundary(bagging)

###Accuracy without bootstrap
bootstrap = False
bagging = BaggingClassifier(
        DecisionTreeClassifier(criterion='entropy', max_depth=max_depth, random_state=0), 
        max_samples=max_samples, bootstrap=bootstrap, n_estimators=n_estimators, random_state=0)
bagging.fit(dataset_train, dataset_labels_train)
print("Accuracy without bootstrap: {}".format(bagging.score(dataset_test, dataset_labels_test)))
plot_decision_boundary(bagging)
```

```python
# imports
import numpy as np
import pandas as pd
from sklearn.utils import shuffle
from xgboost import XGBClassifier
```

```python
# load the data
data_dir = ''

df_train = pd.read_csv(data_dir + 'fashion_mnist_train.csv')
df_test = pd.read_csv(data_dir + 'fashion_mnist_test.csv')

x_train = df_train.drop('label', axis=1).as_matrix().astype(np.uint8)
y_train = df_train['label'].as_matrix().astype(np.uint8)
x_test = df_test.drop('label', axis=1).as_matrix().astype(np.uint8)
y_test = df_test['label'].as_matrix().astype(np.uint8)
```

```python
# Data selection: tee-shirt and ankle boots
np.random.seed(0)
x_train_tb = np.concatenate((x_train[y_train==0], x_train[y_train==9]), axis=0)
y_train_tb = np.concatenate((y_train[y_train==0], y_train[y_train==9]), axis=0)
x_test_tb = np.concatenate((x_test[y_test==0], x_test[y_test==9]), axis=0)
y_test_tb = np.concatenate((y_test[y_test==0], y_test[y_test==9]), axis=0)
x_train_tb, y_train_tb = shuffle(x_train_tb, y_train_tb, random_state=0)
x_test_tb, y_test_tb = shuffle(x_test_tb, y_test_tb, random_state=0)

print("Number of training samples: ", len(x_train_tb))

```

```python
# Fitting and tuning of random forest 
#look for best number of trees
accuracies = []
for n_tree in range(1, 102, 10):
    forest = RandomForestClassifier(criterion='gini', bootstrap=True,
                                    n_estimators=n_tree,
                                    n_jobs=-1, random_state=0)
    forest.fit(x_train_tb, y_train_tb)
    accuracies.append(forest.score(x_test_tb, y_test_tb))

## Plot
fig, ax = plt.subplots()
ax.plot(np.arange(1,102,10), accuracies)
ax.set(xlabel='number of trees', ylabel='accuracy', title='Accuracy for different number of trees')
ax.grid()

#look for best depth
accuracies = []
for depth in range(1, 21):
    forest = RandomForestClassifier(criterion='gini', bootstrap=True,
                                    max_depth=depth, n_estimators=22,
                                    n_jobs=-1, random_state=0)
    forest.fit(x_train_tb, y_train_tb)
    accuracies.append(forest.score(x_test_tb, y_test_tb))

## Plot
fig, ax = plt.subplots()
ax.plot(np.arange(1,21), accuracies)
ax.set(xlabel='depth of trees', ylabel='accuracy', title='Accuracy for different depth of trees')
ax.grid()

```

```python
# Fitting and tuning of Adaboost
#look for best number of trees
accuracies = []
for n_tree in range(1, 82, 10):
    ada_forest = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=1),
                                    n_estimators=n_tree, learning_rate=1, random_state=0)
    ada_forest.fit(x_train_tb, y_train_tb)
    accuracies.append(ada_forest.score(x_test_tb, y_test_tb))

## Plot
fig, ax = plt.subplots()
ax.plot(np.arange(1,82,10), accuracies)
ax.set(xlabel='number of trees', ylabel='accuracy', title='Accuracy for different number of trees')
ax.grid()

#look for best depth
accuracies = []
for depth in range(1, 5):
    ada_forest = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=depth),
                                    n_estimators=40, learning_rate=1, random_state=0)
    ada_forest.fit(x_train_tb, y_train_tb)
    accuracies.append(ada_forest.score(x_test_tb, y_test_tb))

## Plot
fig, ax = plt.subplots()
ax.plot(np.arange(1,5), accuracies)
ax.set(xlabel='depth of trees', ylabel='accuracy', title='Accuracy for different depth of trees')
ax.grid()

```

```python
# Fitting and tuning random forest on all data

#look for best number of trees
accuracies = []
for n_tree in range(1, 82, 10):
    forest = RandomForestClassifier(criterion='gini', bootstrap=True,
                                    n_estimators=n_tree,
                                    n_jobs=-1, random_state=0)
    forest.fit(x_train, y_train)
    accuracies.append(forest.score(x_test, y_test))

## Plot
fig, ax = plt.subplots()
ax.plot(np.arange(1, 82,10), accuracies)
ax.set(xlabel='number of trees', ylabel='accuracy', title='Accuracy for different number of trees')
ax.grid()

#look for best depth
accuracies = []
for depth in range(1, 5):
    forest = RandomForestClassifier(criterion='gini', bootstrap=True,
                                    max_depth=depth, n_estimators=22,
                                    n_jobs=-1, random_state=0)
    forest.fit(x_train, y_train)
    accuracies.append(forest.score(x_test, y_test))

## Plot
fig, ax = plt.subplots()
ax.plot(np.arange(1,21), accuracies)
ax.set(xlabel='depth of trees', ylabel='accuracy', title='Accuracy for different depth of trees')
ax.grid()
```
